```python
### model creation
x = df.drop(columns = ["price"])
y = df.price
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
#### in linear reg problem we want to reach the global minima fast that why we transform the all the data 
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
import pickle
pickle.dump(scaler,open("scaling.pkl","wb"))
```

```python
### model traing
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
#### on which parameter model is trained
lr.get_params()
```

```python
### prediction
reg_pred = lr.predict(x_test)
```

```python
### check the assumption
plt.scatter(y_test,reg_pred)
```

```python
residual = y_test - reg_pred
sns.distplot(residual)
```

```python
plt.scatter(residual,reg_pred)
```

```python
### r2 squre and adjusted r2 
from sklearn.metrics import r2_score
score = r2_score(y_test,reg_pred)
score
```
